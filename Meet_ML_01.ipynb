# **1.1.	Apa itu Kecerdasan Buatan dan Machine Learning?**

Kecerdasan Buatan (Artificial Intelligence atau disingkat AI) adalah kemampuan sistem komputer untuk menyelesaikan tugas yang biasanya memerlukan kecerdasan manusia, seperti pemrosesan bahasa alami, pengenalan gambar, dan pengambilan keputusan. Sistem AI dapat diatur untuk belajar dari data dan pengalaman, sehingga mereka dapat terus meningkatkan kinerja mereka dalam menyelesaikan tugas-tugas ini.

Machine Learning adalah cabang dari Kecerdasan Buatan yang fokus pada pengembangan algoritma komputer yang dapat belajar dari data dan menghasilkan hasil yang berguna tanpa perlu diprogram secara eksplisit. Machine learning dapat digunakan dalam berbagai aplikasi seperti klasifikasi gambar, pengenalan suara, dan analisis data bisnis.

# **1.2. Jenis Data**

**a. Apa itu Data Terstruktur?**

Data Terstruktur adalah jenis data yang terorganisasi dengan baik dan memiliki format yang tetap dan konsisten. Data terstruktur seringkali disimpan dalam basis data relasional dan mudah dianalisis menggunakan perangkat lunak khusus seperti SQL. Contoh data terstruktur adalah data dalam bentuk tabel seperti data pelanggan, data produk, atau data transaksi.

Sumber: Wibowo, A., & Mahardika, G. (2018). Big Data: Konsep dan Aplikasi. Andi Publisher.

**b. Apa itu Data Tidak Terstruktur?**

Data Tidak Terstruktur adalah jenis data yang tidak memiliki format yang terorganisasi dengan baik dan sulit untuk dianalisis menggunakan perangkat lunak konvensional. Data ini bisa berupa teks, suara, video, atau gambar. Contoh data tidak terstruktur adalah dokumen teks, email, postingan media sosial, dan rekaman suara atau video.

Sumber: Puspitasari, N. A., & Akbar, M. A. (2020). Data Science: Panduan Praktis Memahami Data untuk Bisnis dan Teknologi.

# **1.3. Jenis Pembelajaran**

**a. Apa itu Supervised Learning?**

Supervised Learning adalah jenis pembelajaran mesin di mana model pembelajaran mesin belajar dari contoh yang sudah di-annotasi. Data yang digunakan dalam supervised learning sudah memiliki label atau kategori yang jelas sehingga mesin dapat mengenali pola dan relasi antara input dan output. Supervised learning cocok digunakan dalam aplikasi seperti klasifikasi, regresi, dan prediksi.

Sumber: Arifin, Z. (2018). Kecerdasan Buatan (Artificial Intelligence) & Machine Learning: Teori & Implementasi. PT. Gramedia Widiasarana Indonesia.

**b. Apa itu Unsupervised Learning?**

Unsupervised Learning adalah jenis pembelajaran mesin di mana model pembelajaran mesin belajar tanpa adanya label atau kategori yang jelas dalam data. Dalam unsupervised learning, mesin mencoba menemukan pola atau struktur yang tidak terlihat sebelumnya dalam data yang tidak berlabel. Unsupervised learning cocok digunakan dalam aplikasi seperti klasifikasi data, analisis data, dan segmentasi data.

Sumber: Arifin, Z. (2018). Kecerdasan Buatan (Artificial Intelligence) & Machine Learning: Teori & Implementasi. PT. Gramedia Widiasarana Indonesia.

# **1.4. Missing Data pada Statistik** 

Definisi "missing data"
"Missing data" pada statistik merujuk pada data yang hilang atau tidak lengkap dalam suatu dataset. Hal ini dapat terjadi karena beberapa alasan seperti kesalahan dalam pengumpulan data, data yang tidak relevan, atau karena responden menolak untuk memberikan informasi.

Pengaruh "missing data" pada analisis statistik
Adanya "missing data" dapat mempengaruhi analisis statistik dan menghasilkan kesimpulan yang tidak akurat. Contohnya, jika sejumlah besar data hilang pada variabel tertentu, maka analisis statistik dapat menghasilkan estimasi yang bias atau standar kesalahan yang tidak dapat diandalkan.

Metode penanganan "missing data"
Penanganan "missing data" dalam analisis statistik sangat penting dan dapat dilakukan dengan berbagai metode. Beberapa metode yang umum digunakan meliputi:

Penghapusan data yang hilang: Data yang hilang dihapus dari dataset sebelum melakukan analisis statistik. Metode ini dapat menghasilkan dataset yang lebih kecil dan dapat mempengaruhi akurasi analisis.

Imputasi data: Data yang hilang diganti dengan nilai yang dihitung dari data yang tersedia. Ada beberapa metode imputasi yang dapat digunakan seperti imputasi nilai tengah, imputasi regresi, dan imputasi multiple imputation.

Penggunaan model statistik khusus: Ada beberapa model statistik khusus yang dapat digunakan untuk menangani "missing data" seperti model likelihood-based dan model multiple imputation.

Pilihan metode penanganan "missing data"
Pilihan metode penanganan "missing data" tergantung pada sifat data yang hilang dan tujuan analisis statistik yang dilakukan. Misalnya, jika hanya sedikit data yang hilang, maka metode imputasi dapat menjadi pilihan yang baik. Namun, jika sejumlah besar data hilang, maka penghapusan data yang hilang mungkin menjadi pilihan yang lebih baik.

# **1.5. Apa itu library Pandas dan apa itu library Numpy** 

Pandas dan NumPy adalah dua library Python yang sering digunakan dalam analisis data dan ilmu data.

Pandas menyediakan struktur data yang mudah digunakan dan fleksibel seperti DataFrame dan Series, dan memungkinkan pengguna untuk melakukan tugas seperti manipulasi, pengolahan, dan analisis data.

Sementara NumPy adalah library untuk komputasi numerik dalam Python, dengan fitur-fitur seperti array multidimensi, fungsi matematika, dan algoritma linier.

Keduanya digunakan dalam proses analisis data, meskipun dengan fokus yang sedikit berbeda. Pandas lebih fokus pada manipulasi dan analisis data, sedangkan NumPy lebih fokus pada operasi matematika dan numerik.

# **5.	Sebutkan Perbredaan Supervised Learning dan Unsupervised Learning, beserta algoritma dari masing-masing tersebut!**

Perbedaan Supervised Learning dan Unsupervised Learning:
1.	Supervised Learning:<br>
•	Memiliki data latih yang telah diberi label atau target output.
•	Memiliki tujuan untuk memprediksi output yang tepat berdasarkan input yang diberikan.
Contoh algoritma: Linear Regression, Logistic Regression, Decision Tree, Random Forest, Neural Networks, dll.
2.	Unsupervised Learning:
•	Tidak memiliki data latih yang telah diberi label atau target output.
•	Tujuan dari algoritma unsupervised adalah mengelompokkan atau mengklasifikasikan data menjadi kelompok-kelompok tertentu berdasarkan kesamaan atau pola yang ada di dalam data.
Contoh algoritma: K-Means, Hierarchical Clustering, DBSCAN, PCA, dll.
